```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
companies = pd.read_csv("/content/drive/MyDrive/EMP BEN PROJECT DATA/employer_health_plan_dataset.csv")
employees = pd.read_csv("/content/drive/MyDrive/EMP BEN PROJECT DATA/cleaned_employee_workforce_profile.csv")
```

Aggregate statistics for the company benefits plans across industries

Since we are focusing on the tech industry, these are the aggregate statistics for the companies within the tech industry in the dataset

```python
# Filter to Technology industry
tech_companies = companies[companies["industry"] == "Technology"]

# Define Company X
company_x = companies[companies["company_id"] == 22]

company_x
```

Company_22 will be the focus of our benchmark analysis

## Company X Overview

Company X is a growing organization within the Technology industry, employing approximately 572 employees.

## Industry Scope

This analysis focuses specifically on companies within the Technology sector to ensure relevant benchmarking.

Company_22 will be evaluated against peer organizations operating within the same industry.

## Premium Comparison (Technology Industry)

```python
tech_companies.groupby("company_name")["monthly_premium"].mean().sort_values().plot(kind="bar", figsize=(10,5))

plt.axhline(company_x["monthly_premium"].values[0], color='red', linestyle='--', label="Company_22")

plt.title("Monthly Premium Comparison (Technology Industry)")
plt.xlabel("Company")
plt.ylabel("Monthly Premium ($)")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
avg_tech_premium = tech_companies["monthly_premium"].mean()
company_22_premium = company_x["monthly_premium"].values[0]

comparison_df = pd.DataFrame({
    'Company': ['Company_22', 'Average Technology Company'],
    'Monthly Premium': [company_22_premium, avg_tech_premium]
})

plt.figure(figsize=(8, 6))
sns.barplot(x='Company', y='Monthly Premium', data=comparison_df, palette='viridis')
plt.title('Monthly Premium Comparison: Company_22 vs. Average Technology Company')
plt.ylabel('Monthly Premium ($)')
plt.show()
```

Company_22’s premium level is compared directly against other Technology firms, providing a more relevant benchmark than cross-industry comparisons. As you can see Company_22's monthly premium registers a few is higher than the Average monthly premium of the Technology industry in this dataset.

This suggests a more generous benefits offering which may enhance employee satisfaction but increase employer cost exposure.

## Employer Contribution Benchmark



```python
tech_companies.groupby("company_name")["employer_contribution_pct"].mean().sort_values().plot(kind="bar", figsize=(10,5))

plt.axhline(company_x["employer_contribution_pct"].values[0], color='red', linestyle='--')

plt.title("Employer Contribution % (Technology Industry)")
plt.show()
```

```python
avg_tech_employee_contribution = tech_companies["employer_contribution_pct"].mean()
company_22_employee_contribution = company_x["employer_contribution_pct"].values[0]

comparison_df = pd.DataFrame({
    'Company': ['Company_22', 'Average Technology Company'],
    'Employer Contribution': [company_22_employee_contribution, avg_tech_employee_contribution]
})

plt.figure(figsize=(8, 6))
sns.barplot(x='Company', y='Employer Contribution', data=comparison_df, palette='viridis')
plt.title('Employer Contribution Comparison: Company_22 vs. Average Technology Company')
plt.ylabel('Employer Contribution (%)')
plt.show()
```

As previously observed, Company_22 may have higher monthly premiums than the average tech company but they also report higher employer contributions than the average tech company lowering the cost of for their employees as well.  

This also strengthens the implication that company_22 has a generous benefit offering and seeks to offer employees high level benefits and is willing to cover the costs.

## Deductible Structure Analysis
